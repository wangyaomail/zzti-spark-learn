```python
filename = '负面词.dict'
allwords = []
with open(filename, 'r', encoding='utf-8') as f:
    for line in f.readlines():
        toks = line.rstrip()
        if len(toks) > 1:
            print(line)
            allwords.append(line)
allwords.sort()
```

```python
with open(filename, 'w', encoding='utf-8') as f:
    for line in allwords:
        f.write(line)
```
